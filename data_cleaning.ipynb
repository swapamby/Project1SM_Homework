```python
# DATA CLEANING

# Dependencies

import csv
import json

with open('movie_data_raw.txt') as json_file:
    data = json.load(json_file)
```

```python
# Filtering out movies: with no data (bad API call), year doesn't match (misplaced), year has no data
# Converting BoxOffice Data type from integer to string

for year in list(data.keys()):
    movies = list(data[year].keys())
    for movie in movies:
        if data[year][movie]['Response'] == 'False':
            data[year].pop(movie)
            print(f'POPPED: no response data - {year} - {movie}')
            continue
        elif data[year][movie]['Year'] == 'N/A':
            data[year].pop(movie)
            print(f'POPPED: no year data - year: {year} - {movie}')
            continue
        elif data[year][movie]['Year'] != year and data[year][movie]['Year'] != str(int(year)-1) and data[year][movie]['Year'] != str(int(year)+1): 
            data[year].pop(movie)
            print(f'POPPED: year does not match - {year} - {movie}')
            continue
        elif data[year][movie]['BoxOffice'] == 'N/A':
            data[year].pop(movie)
            print(f'POPPED: no BoxOffice data - year: {year} - {movie}')
            continue
        else:
            try:
                BO = data[year][movie]['BoxOffice'][1:].split(',')
                BO_val = ''
                for i in range(len(BO)):
                    BO_val = BO_val + BO[i]
                data[year][movie]['BoxOffice'] = int(BO_val)
            except Exception as e:
                data[year].pop(movie)
                print(f'POPPED: ???? - {year} - {movie} - {e}')
```

```python
# Check to see how things went

for year in list(data.keys()):
    movies = list(data[year].keys())
    print(f"{year} has {len(movies)} movies")
    for movie in movies:
        if data[year][movie]['Response'] == 'False':
            print('HELP what movie?')
        elif data[year][movie]['Year'] == 'N/A' or data[year][movie]['Year'] != year:
            print("HELPPP my movie doesn't have a year")
        elif type(data[year][movie]['BoxOffice']) == 'str':
            print('HEEEELLLLLPPPPP with Box Office stuff')
        else: continue
```

```python
with open('movie_data.txt', 'w') as outfile:
    json.dump(data, outfile)
```
